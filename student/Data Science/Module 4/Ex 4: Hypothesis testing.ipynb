```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
tips =sns.load_dataset('tips')
tips.head()
```

```python
tips.describe()
```

```python
# Check for nulls to ensure data readiness
print(tips.isnull().sum())
```

Here’s a description of each column in the dataset:

- total_bill: The total bill amount, including the - cost of food and drinks.
- tip: The tip amount given by the customer.
- sex: The gender of the customer (e.g., Male or Female).
- smoker: Whether the customer is a smoker or not (e.g., Yes or No).
- day: The day of the week when the transaction occurred (e.g., Sun, Sat, Thu, etc.).
- time: The time of day when the transaction occurred, typically categorized as Lunch or Dinner.
- size: The size of the party or group of customers.


#### H1​: smoking is associated with time of visit

```python
sns.catplot(data=tips, x='time', hue='smoker', col='day', kind='count', height=4, aspect=0.8)
plt.Figure(figsize=(50,35))
plt.subplots_adjust(top=0.85)
plt.suptitle('Smoking Distribution by time based on the day')
plt.show()
```

#### H2​: the bigger the group the higher the tip

```python
plt.Figure(figsize=(20,20))
sns.regplot(data=tips,x='size', y='tip', scatter_kws={'alpha':0.5})
plt.title('Relationship between Group Size and Tip Amount')
plt.show()
```

#### H3​: group size is different based on the time of visit

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=tips, x='time', y='size', color='skyblue')
plt.title('Comparison of Group Sizes: Lunch vs Dinner')
plt.show()
```

#### H4H4​: (… come up with a hypothesis of your own …)

** H4: Total bill amounts vary significantly by gender, which acts as a primary driver for the differences in total tips. **

##### 1.Do men pay higher bills than women (or vice versa)?

```python
sns.boxenplot(data=tips, x='sex', y= 'total_bill',hue='sex', palette={"Male":"deepskyblue", "Female": "hotpink"},legend=False)
plt.title('Total Bill Distribution by Gender')
```

#### 2.Does this difference affect tips?

```python

custom_colors = {"Male": "deepskyblue", "Female": "hotpink"}

sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex', palette=custom_colors, height=6)
plt.title('Relationship between Total Bill and Tip by Gender')
```

#### Finally, analyze if size (party size) is a confounder. That is, does a larger party cause a higher tip, or simply a higher bill which then leads to a higher tip?

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='size', palette='viridis')
```

```python
tips['tip_pct'] = (tips['tip'] / tips['total_bill']) * 100
sns.barplot(data=tips, x='size', y='tip_pct', palette='viridis',legend=False, errorbar=None)
plt.title('Average Tip Percentage by Group Size')
plt.ylabel('Average Tip Percentage (%)')
plt.xlabel('Group Size (Number of People)')
plt.show()
```

#### While group size correlates with higher bills, regression indicates it is not a direct driver of tips. Instead, it acts as a **confounder** that influences tipping indirectly through increased total spending."

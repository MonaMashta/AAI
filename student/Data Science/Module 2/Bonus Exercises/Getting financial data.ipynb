```python
df_tesla = yf.download("TSLA", start=start, end=end)
df_tesla["stock"] = "TSLA"


df_ibm = yf.download("IBM", start=start, end=end)
df_ibm["stock"] = "IBM"


df_microsoft = yf.download("MSFT", start=start, end=end)
df_microsoft["stock"] = "MSFT"


df_tesla.head()
```

### Step 7. Combine the four separate dataFrames into one combined dataFrame df that holds the information for all four stocks

```python
frames = [df_apple, df_tesla, df_ibm, df_microsoft]

df = pd.concat(frames)
df

```

### Step 8. Shift the stock column into the index (making it a multi-level index consisting of the ticker symbol and the date).

```python
df.set_index(keys="stock", append=True, inplace=True)
df
```

### Step 7. Create a dataFrame called vol, with the volume values.

```python
vol = df['Volume']

vol = pd.DataFrame(vol)

vol
```

### Step 8. Aggregate the data of volume to weekly.

Hint: Be careful to not sum data from the same week of 2015 and other years.

```python
vol_flat = vol.reset_index()

vol_flat['year'] = vol_flat['Date'].dt.year
vol_flat['week'] = vol_flat['Date'].dt.isocalendar().week

vol_weekly = vol_flat.groupby(['year', 'week']).sum(numeric_only=True)

vol_weekly.head(15)
```

### Step 9. Find all the volume traded in the year of 2015

```python
total_volume_2015 = vol_weekly.loc[2015].sum().sum()

print(f"Total volume traded in 2015: {total_volume_2015}")
```

```python
vol_2015_display = vol_weekly.loc[2015].sum().to_frame().T

vol_2015_display.index = [2015]
vol_2015_display.index.name = 'year'

vol_2015_display
```

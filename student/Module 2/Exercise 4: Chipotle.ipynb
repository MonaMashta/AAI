Step 1. Import the necessary libraries

```python
import pandas as pd
```

Step 2. Import the dataset from this address.

Step 3. Assign it to a variable called chipo.

```python
url= "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"

chipo = pd.read_csv(url,sep="\t")
```

Step 4. How many products cost more than $10.00?

```python
chipo.head(1)
```

```python

chipo["item_price"]= chipo.item_price.apply(lambda x: (float(x[1:-1])))
chipo.head()
```

```python
over10= chipo[chipo["item_price"]>10.00]
print(over10['item_name'].nunique())
```

Step 5. What is the price of each item?


print a data frame with only three columns item_name choice_description and product_price

```python
menu= chipo[['item_name','choice_description','item_price']].sort_values(by= 'item_price', ascending= False).drop_duplicates()
menu
```

Step 6. Sort by the name of the item

```python
chipo.sort_values(by= 'item_name', ascending=False)
```

Step 8. How many times was a Veggie Salad Bowl ordered?

```python
qua = chipo[chipo['item_name']=="Veggie Salad Bowl"]['quantity'].sum()
print(qua)
```

Step 9. How many times did someone order more than one Canned Soda?

```python
sodas= ((chipo['item_name'] == "Canned Soda")& (chipo['quantity'] > 1)).sum()
print(sodas)
```
